```python
import numpy as np
import pandas as pd
dataset = pd. read_csv ("train.csv")
dataset.head()
```

```python
dataset.shape
```

```python
dataset.describe()
dataset["label"] = dataset['label'].astype('category')
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
#Remove the special characters, numbers etc.
#Lemmatize the text
from nltk.stem import WordNetLemmatizer
from nltk import tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
import re
```

```python
import nltk
nltk.download('wordnet')
```

```python
import nltk
nltk.download('omw-1.4')
```

```python

dataset['text_lem'] = [''.join([WordNetLemmatizer().lemmatize(re.sub('[^A-Za-z]',' ',text)) for text in lis]) for lis in dataset['tweet']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split (dataset['text_lem'], dataset['label'])
```

```python
vect = TfidfVectorizer(ngram_range= (1,4)).fit(X_train)
vect_transformed_X_train =vect.transform(X_train)
vect_transformed_X_test= vect.transform(X_test)
print (vect_transformed_X_train)
```

```python
print(vect_transformed_X_test)
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import f1_score
#F1 score is used as an evaluation measure when the data is skewed like in this case, where the number of hate speech tweets are very less, accuracy cannot
#be relied upon.
```

```python
#Support Vector Machine
modelSVC = SVC(C=100).fit(vect_transformed_X_train,y_train)
modelSVC
```

```python
predictionsSVC= modelSVC.predict(vect_transformed_X_test)
sum(predictionsSVC==1), len(y_test), f1_score(y_test, predictionsSVC)
```

```python
#Logistic Regression
modelLR= LogisticRegression(C=100).fit(vect_transformed_X_train,y_train)
```

```python
predictionsLR= modelLR.predict(vect_transformed_X_test)
sum(predictionsLR==1), len(y_test), f1_score(y_test, predictionsLR)
```

```python
#Navie Bayes
```

```python
modelNB = MultinomialNB(alpha=1.7).fit(vect_transformed_X_train,y_train)
```

```python
predictionsNB= modelNB.predict(vect_transformed_X_test)
sum(predictionsNB==1), len(y_test), f1_score(y_test, predictionsNB)
```

```python
#Random Forest
modelRF = RandomForestClassifier(n_estimators=20).fit(vect_transformed_X_train,y_train)
```

```python
predictionsRF =modelRF.predict(vect_transformed_X_test)
sum(predictionsRF==1), len(y_test), f1_score(y_test, predictionsRF)
```

```python
#Stochastic Gradient Descent
```

```python
modelSGD = SGDClassifier (loss='hinge', penalty='l2', alpha=1e-3).fit(vect_transformed_X_train,y_train)
```

```python
predictionsSGD =modelSGD.predict(vect_transformed_X_test)
sum(predictionsSGD==1), len(y_test), f1_score(y_test, predictionsSGD)
```

```python
#Based on all the above models trained we conclude that the Logistic regression is the better model amoung them
#so we use this model as our final model.
```

```python
from textblob import TextBlob
def getSubjectivity(tweet) :
    return TextBlob(tweet).sentiment.subjectivity
def getPolarity (tweet) :
    return TextBlob(tweet).sentiment.polarity
def getAnalysis (score) :
    return 'Neutral' if (score == 0) else ('Negative' if (score < 0) else 'Positive')
```

```python
#polarity output
```

```python
test['Subjectivity']= test['tweet'].apply(getSubjectivity)
test['Polarity']= test['tweet'].apply(getPolarity)
test['Analysis'] = test['Polarity'].apply(getAnalysis)
```
